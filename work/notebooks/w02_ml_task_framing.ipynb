## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My lane is Refresh / Content Opportunity Scoring.

This is a ranking task because the goal is to prioritize content pages based on their refresh opportunity. Content teams often manage thousands of pages and cannot review all of them manually. A ranking system helps identify which pages should be reviewed first by ordering them from highest to lowest opportunity.

The output supports the decision of which content pages should be refreshed first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

There is no direct label called "needs refresh" in the dataset, so I would use a proxy target.

I would use trend_pct and trend_direction as indicators of refresh opportunity. Pages with large negative trends are likely losing visibility or engagement and may benefit from a refresh.

For example, pages with trend_pct below -30% could be considered high refresh opportunity, while pages with stable or positive trends could be considered lower priority.

This target is based on observed performance signals combined with a defined business rule.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The primary success metric is Precision@K.

The content team can only refresh a limited number of pages, so the model should place the highest-opportunity pages at the top of the ranking. Precision@K measures how many of the top recommended pages truly deserve review.

A good model would consistently identify pages with meaningful performance decline among its highest-ranked recommendations.

I would also monitor NDCG to evaluate the overall ranking quality.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule such as "refresh pages older than one year" is too simplistic.

Refresh opportunity depends on many factors, including search volume, competition, CTR, average position, engagement rate, scroll rate, and trend percentage. Some old pages continue to perform well and do not need updates, while some newer pages may be losing traffic and engagement.

Machine learning can combine multiple signals and learn patterns that are difficult to capture with a single if-statement. This allows content teams to prioritize refresh efforts more effectively and make more data-driven decisions.
