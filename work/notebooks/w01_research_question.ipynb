I chose the **Refresh / Content Opportunity Scoring** lane.

Many websites contain hundreds or thousands of pages, but content teams have limited time and resources. Instead of reviewing every page manually, a scoring system can help prioritize which pages should be reviewed first.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

### Search Question

Which webpages should be reviewed first for content refresh or optimization?

### Unit of Analysis

One webpage (one content page).

### Output

A ranked review queue or priority score for each webpage.

### Decision

Which pages should be selected for review first.

### Action

review selected pages and decide whether to:
- Update outdated information
- Improve titles or descriptions
- Expand content
- Improve engagement signals
- Monitor performance more closely

### Cost of a Wrong Recommendation

False Positive:
A page is recommended for review even though it does not need attention, causing time and effort to be wasted.

False Negative:
An important page that needs attention is missed, potentially allowing traffic, engagement, or visibility to continue declining.
